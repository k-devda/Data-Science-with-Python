```python
# Accessing using Custom Index 

import pandas as pd 

data = pd.Series (
        [80,78,98,90],
    index=["Khushi","Riya","Shalin","Jiya"],
    name="Marks"
)

print(data["Khushi"]) #Custom_Index 
print(data ["Jiya"]) #Custom_Index

#Access through Postion  Value 

print(data.iloc[0])
print(data.iloc[3])

# Only WantsValues No index then use these 
print(data.values)

# Want only index not  Values then use these function 
print(data.index)

#these Function is used when u want the data's Type 
print(data.dtype)

#if you want the size How many Elements are there ?
print(data.size)

#Shape (Returns the Dimensions )
print(data.shape)

#these function will return the name of the series not coloumn 
print(data.name)
print("Sorting by the values in asscending order ")
#Sorting by the values in asscending order 
print(data.sort_values(ascending=True))
print("Sorting by the values in descending order ")
print(data.sort_values(ascending=False))

print("Sorting by the Index values in asscending order ")
#Sorting by the values in asscending order 
print(data.sort_index(ascending=True))
print("Sorting by the Index  values in descending order ")
print(data.sort_index(ascending=False))
```

```python
import pandas as pd 

mobile = pd.Series (

    [20000,50000,10000],
    index=["Samsung","Apple","Realme"],
    name="Price"
)

mobile

#Print all values.
print(mobile.values)

#Print all indexes.
print(mobile.index)

#Print the data type.
print(mobile.dtype)

#Print the size.
print(mobile.size)


#Print the shape.
print(mobile.shape)

#Print the name.
print(mobile.name)
```

```python
#now the methods 

import pandas as pd

salary = pd.Series(
    [25000, 30000, 28000, 35000, 30000]
)

#Total salary
print(salary.sum())

#Average salary
print(salary.mean())

#Highest salary
print(salary.max())

#Lowest salary
print(salary.min())

#Number of salaries
print(salary.count())

#Median salary
print(salary.median())

#Mode salary
print(salary.mode())
```
